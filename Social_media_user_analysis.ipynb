# 1. Comparsion Accoding Ages 

```python
sns.set(rc={'figure.figsize':(20,10)}) 
ax = sns.countplot(x='age' ,data=df )
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Comparsion between age and gender :
df.groupby(['gender'],as_index=False)['age'].sum().sort_values(by='age',ascending=False)
```

## Result : From above graph we see that the most of Age male using social media under 18 - 64 year
